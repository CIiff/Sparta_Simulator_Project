```python
import numpy as np
import pandas as pd
import random
import scipy.stats as stats

def trainee_generator(min_trainees,max_trainees):
    new_train_mean = (min_trainees + max_trainees)/2.0
    new_train_stdev = (new_train_mean - min_trainees)/3.0
    num_new_trainees = float(stats.truncnorm.rvs(
        (min_trainees-new_train_mean)/new_train_stdev,
        (max_trainees-new_train_mean)/new_train_stdev,
        loc=new_train_mean, scale=new_train_stdev, size=1))
    return round(num_new_trainees)

print('done')
```

```python
trainee_df = pd.DataFrame(columns=["Assigned centre ID", "Course type", "Start month", "Stop month","Status"])
courses = ["Data", "DevOps", "C#", "Java", "Business"]
num_new_trainees = trainee_generator(20,30)
print(num_new_trainees)
for trainee in range(num_new_trainees):
    
    row_data = {"Course type": random.choice(courses), "Assigned centre ID": "None",
                        "Start month": 0, "Stop month": 0, "Status": "Waiting"}
    trainee_df = trainee_df.append(row_data, ignore_index=True)
print('A+++')
print(trainee_df.iloc[0:3,1])
print('B+++',trainee_df.iloc[0,:]['Course type'])
print('C+++',trainee_df.index[trainee_df['Course type'] == 'Java'].tolist())
print('D+++',trainee_df['Course type'].iloc[0])
for i, row in trainee_df.iterrows(): # i = index, row= all row info
    print(i,row['Course type'])
    if i in [5,7,14]:
        row['Course type'] = 'SomeOtherCourse'
trainee_df
```
